```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

def get_html(date, team, opp) -> str:
    formatted_date = date.replace('-','')
    url = f'https://www.baseball-reference.com/boxes/{team}/{team}{formatted_date}0.shtml'
    response = requests.get(url=url)
    if response.status_code == 400:
        url = f'https://www.baseball-reference.com/boxes/{opp}/{opp}{formatted_date}0.shtml'
        response = requests.get(url=url)
    data_html = response.text
    return data_html

def format_team_name(soup) -> str:
    pitch_team = soup.tbody.tr.find_all("td")[1].get_text()
    team_name = pitch_team.replace(' ','')
    return team_name

def get_pitching_table(data_html, team_name_re) -> pd.DataFrame:
    t = re.findall(f'div class=\"table_container\" id=\"div_{team_name_re}[\s\S]+?</table>', data_html)
    df = pd.read_html(t[1])
    return df

def team_name_lookup(team_code) -> str:
    lookup = {
        'ARI':'Arizona Diamondbacks'
    }
    team_name = lookup[team_code]
    team_name.replace(' ','')
    return team_name

#data_html = get_html('2014-03-22', 'LAD')
```
